# Step 4: Implementing the Dense Layer Class in Python

Firstly, we'll start with a data generation function that we'll use from now on to generate challenging, non-linear data for our neural network to detect patterns in.

Eventually, our network will be able to take some spiral data and separate it into individual spirals, as shown below: 



```python
from nnfs.datasets import spiral_data
import nnfs
import numpy as np
import matplotlib.pyplot as plt
nnfs.init()

# note: x contains both x and y coordinates of the data, y contains the class info
x, y = spiral_data(samples=100, classes=3)
plt.scatter(x[:, 0], x[:, 1])
plt.show()
```

Dense Layer class:

```python
class DenseLayer:
    def __init__(self, n_inputs, n_neurons):
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))
    
    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases

dense1 = DenseLayer(2, 3)
dense1.forward(x)
dense1.output[:5]
```
